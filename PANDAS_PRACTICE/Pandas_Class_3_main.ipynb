```python
# chunksize read_csv()


data=pd.read_csv('iris.csv')
data
```

```python
data=pd.read_csv('iris.csv',chunksize=10)
```

```python
data
```

```python
# to print chunksize let we use for loop

for chunk in data:
    print(chunk)
```

```python
# to know the total dataset using chunk

data=pd.read_csv('iris.csv',chunksize=10)
t_len=0
for chunk in data:
    t_len += len(chunk)
print(t_len)
```

```python
# in general, if we use chunksize, the output wil returns as textfile reader dataset
data=pd.read_csv('iris.csv',chunksize=10)
data
```

```python
# to change from textfile reader dataset to normal dataframe

df=pd.concat(data)
df
```
