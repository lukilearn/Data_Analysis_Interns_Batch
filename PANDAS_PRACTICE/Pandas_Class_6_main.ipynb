# let's take some previous example and do some operation:

```python
d={'City':['Ahmedabad','Baroda','Anand',
           'Ahmedabad','Baroda','Anand'],
   'Temp':[40,35,32,45,37,34],
   'Humidity':[32,34,23,24,27,32]}
data=pd.DataFrame(d)
data
```

# 1. Display Average Temperature

```python
data['Temp'].mean()
```

# 2. Display Average Temperature CityWise.



```python
grp=data.groupby('City').mean()
grp
```

```python
grp=data.groupby('City')['Temp'].mean()
grp
```

# 3. Display All the Groups [Iterate Over the Groups]

```python
grp=data.groupby('City')
grp
```

```python
for key,value in grp:
    print(key)
    print(value)
```

# 4. Display the First Entry in All the Groups Formed

```python
grp=data.groupby('City')
grp.first()
```

# 5. Display the Last Entry in all the Groups Formed

```python
grp.last()
```

# 6. Count The Number of Entries in Each Group


```python
grp.size()
```

# 7. Display A specific Group out of all the Groups


```python
grp=data.groupby('City')
grp.get_group('Ahmedabad')
```

# 8. Apply Functions to Groupby Groups [like min,max,and mean etc.,]

```python
grp=data.groupby('City')
grp.min()
```

```python
grp.max()
```

```python
grp.mean()
```

# 9. Display Average Temperature Citywise with Prefix Mean_Temp_.

```python
grp=data.groupby('City')['Temp'].mean()
.mean().add_prefix('Mean_Temp_')
grp
```

```python
grp=data.groupby('City')['Temp'] \
.mean().add_prefix('Mean_Temp_')
grp
```

# 10. Display Descriptive Statistics



```python
grp=data.groupby('City')
grp.describe()
```

# 11. Perform Statistical Computations Using agg() Function

```python
grp=data.groupby('City')
grp.agg(['min','max','mean'])
```

```python
grp=data.groupby('City').agg(['min','max','mean'])
grp
```

# 12. create a Function to get the statistics Groupwise.

```python
def fun(grp):
    return {'Min':grp.min(),'Mean':grp.mean()}
```

```python
grp=data.groupby('City').apply(fun)
```

```python
grp
```

# Merge Function in Pandas

```python
data1={'Key':['K0','K1','K2','K3','K4'],
       'Name':['Priyang','Aadhya','Vedhant','Prashav','Krisna']}
data2={'Key':['K0','K1','K2','K3','K4'],
       'Marks':[98,99,89,85,87]}

df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(left=df1,right=df2,on="Key")
```

```python
pd.merge(df1,df2,on="Key")
```

```python
# change the position of the dataframe

pd.merge(df2,df1,on="Key")
```

```python
pd.merge(df1,df2)   #default consider automatically take columns in both the dataframe
```

```python
data1={'Key1':['K0','K1','K2','K3','K4'],
       'Key2':['K0','K1','K0','K1','K2'],
       'Name':['Priyang','Aadhya','Vedhant','Prashav','Krisna']}
data2={'Key1':['K0','K1','K2','K3','K4'],
       'Key2':['K0','K1','K0','K1','K2'],
       'Marks':[98,99,89,85,87]}

df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on=['Key1','Key2'])
```
