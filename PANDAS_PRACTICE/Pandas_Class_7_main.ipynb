# Filter DataFrame using & | ! (AND, OR, NOT)

```python
mask1=df["Gender"]=="Male"

mask2=df["Team"]=="Marketing"

df[mask1 & mask2]
```

```python
mask3 = df["Senior Management"]

mask4 = df["Start Date"] < "1990-01-01"

df[mask3 | mask4]             #  |--->> pipe operator
```

```python
mask5 = df["First Name"] == "Robert"

mask6 = df["Team"] == "Client Services"

mask7 = df["Start Date"] > "2016-06-01"

df[(mask5 & mask6) | mask7]
```

```python
 df.head()
```

# join Method

Datafram join is a convenient method for combining the columns of two potentially 
differently-indexed

```python
import pandas as pd

df1=pd.DataFrame({'A':[1,2,3],
                  'B':[10,20,30]})

df2=pd.DataFrame({'C':[1,2,3],
                  'D':[10,20,30]})


```

```python
display(df1,df2)
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```

```python
# Suppose index value of both the dataframe is different means?

df1=pd.DataFrame({'A':[1,2,3],
                  'B':[10,20,30]},
                   index=['a','b','c'])


df2=pd.DataFrame({'C':[4,5,6],
                  'D':[40,50,60]})

```

```python
display(df1,df2)
```

```python
df1.join(df2)    # join operation wil not work if both the dataframe index is different
```

```python
df1=pd.DataFrame({'A':[1,2,3],
                  'B':[10,20,30]},
                   index=['a','b','c'])


df2=pd.DataFrame({'C':[4,5,6],
                  'D':[40,50,60]},
                   index=['a','b','c'])
```

```python
display(df1,df2)
```

```python
df1.join(df2)
```

```python
df1=pd.DataFrame({'A':[1,2,3],
                  'B':[10,20,30]},
                   index=['a','b','c'])


df2=pd.DataFrame({'C':[4,5],
                  'D':[40,50]},
                   index=['a','b'])
```

```python
display(df1,df2)
```

```python
df1.join(df2)
```

```python
df1.join(df2, how='inner')
```

```python
df1.join(df2, how='outer')
```

```python
df1.join(df2, how='right')
```

```python
# suppose if both the dataframe column is name what wil happen???

df1=pd.DataFrame({'A':[1,2,3],
                  'B':[10,20,30]},
                   index=['a','b','c'])


df2=pd.DataFrame({'A':[4,5,6],
                  'D':[40,50,60]},
                   index=['a','b','c'])
```

```python
df1.join(df2)
```

```python
# suppose overlapping column to define some new name for that column

df1.join(df2,lsuffix='_A')
```

```python
df1.join(df2,rsuffix='_A')
```
