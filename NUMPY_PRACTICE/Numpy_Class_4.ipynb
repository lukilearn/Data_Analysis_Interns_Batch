# Ravel vs Flatten vs reshape(-1)


```python
# ravel ---->>> multiple dimension array to convert as single dimension array but it wil consider as view() concept
# suppose if we change any value in x that wil reflect in y also


x=np.arange(6).reshape((3,2))
```

```python
x
```

```python
y=x.ravel()   # converting from 2D to 1D array
```

```python
x
```

```python
y            # converted from 2D array
```

```python
y[1]=999
```

```python
y            # 1D array after value is assigned
```

```python
x            # 2D array after reflected values from 'y' 1D array
```

```python
# flatten() ---->>> multiple dimension array to convert as single dimension array but it wil consider as copy() concept
# suppose if we change any value in x that wil  not reflect in y also
```

```python
x=np.arange(6).reshape((3,2))
```

```python
y=x.flatten()
```

```python
x
```

```python
y
```

```python
y[1]=111
```

```python
x
```

```python
y
```

```python
# reshape(-1) ---->>> any multidimensional array to 1D array


x=np.arange(6).reshape((3,2))
```

```python
y=x.reshape(-1)
```

```python
x
```

```python
y
```

```python
y[1]=888
```

```python
y
```
