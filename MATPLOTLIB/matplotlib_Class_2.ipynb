# plotting Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ["Python","R","AI","ML","DS"]
class1_students=[30,10,20,25,10]
class2_students=[40,5,20,20,10]
class3_students=[35,5,30,15,15]
```

```python
plt.bar(classes,class1_students)
```

```python
# to represent horizontal bar chart

plt.barh(classes,class1_students)
```

```python
# to give bar width size

plt.bar(classes,class1_students,width=0.2)
```

```python
# to align the line of each bar chart

plt.bar(classes,class1_students,width=0.2,align="edge")
```

```python
# to change the color of bar chart

plt.bar(classes,class1_students,width=0.2,align="edge",color="y")
```

```python
# to give edge color of each bar chart

plt.bar(classes,class1_students,width=0.2,align="edge",color="y",
        edgecolor="black")
```

```python
# linewidth access in edgecolor for each bar chart

plt.bar(classes,class1_students,width=0.2,align="edge",color="y",
        edgecolor="black",linewidth=5)

```

```python
# alpha parameter using bar chart

plt.bar(classes,class1_students,width=0.2,align="edge",color="y",
        edgecolor="black",linewidth=5,alpha=0.5)
```

```python
# linestyle parameter using bar chart
plt.bar(classes,class1_students,width=0.2,align="edge",color="y",
        edgecolor="black",linewidth=5,alpha=1,linestyle="--")
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color="y",
        edgecolor="black",linewidth=5,alpha=1,linestyle="--",
        label="class 1 Student")
plt.legend()
```

```python
# to hide the graph

plt.bar(classes,class1_students,width=0.2,align="edge",color="y",
        edgecolor="black",linewidth=5,alpha=1,linestyle="--",
        label="class 1 Student",visible=False)
plt.legend()
```

```python
plt.figure(figsize=(16,8))

plt.bar(classes,class1_students,width=0.2,align="edge",color="y",
        edgecolor="black",linewidth=5,alpha=1,linestyle="--",
        label="class 1 Student",visible=True)
```

```python
plt.barh(classes,class1_students,width=0.2,align="edge",color="y",
        edgecolor="black",linewidth=5,alpha=1,linestyle="--",
        label="class 1 Student")
```

```python
plt.barh(classes,class1_students,align="edge",color="y",
        edgecolor="black",linewidth=5,alpha=1,linestyle="--",
        label="class 1 Student")
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color="y",
        edgecolor="black",linewidth=5,alpha=1,linestyle="--",
        label="class 1 Student")
plt.title("Bar Chart of Student Details")
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.show()
```

```python
plt.bar(classes,class1_students,width=0.2,color="b",label="class 1 Student")

plt.bar(classes,class2_students,width=0.2,color="g",label="class 2 Student")
plt.title("Bar Chart of Student Details")
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.show()
```

```python

plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))

width=0.2
plt.bar(classes_index,class1_students,width,color="b",label="class 1 Student")

plt.bar(classes_index,class2_students,width,color="g",label="class 2 Student")
plt.title("Bar Chart of Student Details")
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.show()
```

```python

plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))

width=0.2
plt.bar(classes_index,class1_students,width,color="b",label="class 1 Student")

plt.bar(classes_index+width,class2_students,width,color="g",label="class 2 Student")
plt.title("Bar Chart of Student Details")
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))

width=0.2
plt.bar(classes_index,class1_students,width,color="b",label="class 1 Student")

plt.bar(classes_index+width,class2_students,width,color="g",label="class 2 Student")
plt.xticks(classes_index+width,classes)
plt.title("Bar Chart of Student Details")
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))

width=0.2
plt.bar(classes_index,class1_students,width,color="b",label="class 1 Student")

plt.bar(classes_index+width,class2_students,width,color="g",label="class 2 Student")
plt.xticks(classes_index+width,classes,rotation=20)
plt.title("Bar Chart of Student Details")
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))

width=0.2
plt.bar(classes_index,class1_students,width,color="b",label="class 1 Student")

plt.bar(classes_index+width,class2_students,width,color="g",label="class 2 Student")

plt.bar(classes_index+width,class3_students,width,color="y",label="class 3 Student")


plt.xticks(classes_index+width,classes,rotation=20)
plt.title("Bar Chart of Student Details")
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))

width=0.2
plt.bar(classes_index,class1_students,width,color="b",label="class 1 Student")

plt.bar(classes_index+width,class2_students,width,color="g",label="class 2 Student")

plt.bar(classes_index+width+width,class3_students,width,color="y",label="class 3 Student")


plt.xticks(classes_index+width,classes,rotation=20)
plt.title("Bar Chart of Student Details")
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.show()
```
