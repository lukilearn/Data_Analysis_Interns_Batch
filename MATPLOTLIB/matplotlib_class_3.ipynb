```python
df_google_play_store_apps=pd.read_csv('googleplaystore.csv',nrows=1000)
```

```python
df_google_play_store_apps.shape
```

```python
x=df_google_play_store_apps["Rating"]
y=df_google_play_store_apps["Reviews"]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c="r")
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.scatter(x,y,c="r",marker="*")
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))

plt.scatter(x,y,c="r",marker="*",s=100)
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))

plt.scatter(x,y,c="r",marker="*",s=100,alpha=0.5)
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))

plt.scatter(x,y,c="r",marker="*",s=100,alpha=0.5,linewidths=10)
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))

plt.scatter(x,y,c="r",marker="*",s=100,alpha=0.5,linewidths=10,edgecolors="g")
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))

plt.scatter(x,y,c="r",s=100,alpha=0.5,linewidths=10,edgecolors="g",verts=">")  # verts is similar to marker
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))

plt.scatter(x,y,c="r",s=100,marker="*",alpha=0.5,linewidths=10,edgecolors="g")

plt.scatter(x,df_google_play_store_apps["Installs"],c="blue",s=100,marker="o",alpha=0.5,linewidths=10,edgecolors="r")


plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews & Installs")
```
