# Histogram Plot using Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import random
```

```python
ml_students_age=np.random.randint(18,45,(100))
py_students_age=np.random.randint(15,40,(100))
```

```python
print(ml_students_age)
print(py_students_age)
```

```python
# General show histogram

plt.hist(ml_students_age)
plt.show()
```

```python
#shows xlabel,ylabel and title

plt.hist(ml_students_age)
plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.show()
```

```python
# added bins values in the hist()

bins=
plt.hist(ml_students_age,bins)
plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.show()
```

```python
# adding rwidth parameter in the hist()

bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8)
plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.show()
```

```python
# adding histtype="step" in the hist()

bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="step")
plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.show()
```

```python
# adding histtype="bar" in the hist()

bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar")
plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.show()
```

```python
# adding orientation="horizantal" parameter in hist()
 
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",
        orientation="horizontal")
plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.show()
```

```python
# adding color in the hist()

bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",
        orientation="horizontal",color="red")
plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.show()
```

```python
# adding label parameter in the hist()

bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",
        orientation="horizontal",color="blue",label="ML Student")
plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.show()
```

```python
# adding label parameter in the hist()

bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",
        orientation="horizontal",color="blue",label="ML Student")
plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

```python
# adding another histogram in to this


bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",
        orientation="horizontal",color="red",label="ML Student")

plt.hist(py_students_age,bins,rwidth=0.8,histtype="bar",
        orientation="horizontal",color="y",label="PY Student")

plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",
        orientation="vertical",color="red",label="ML Student")

plt.hist(py_students_age,bins,rwidth=0.8,histtype="bar",
        orientation="horizontal",color="y",label="PY Student")

plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",
        orientation="vertical",color="red",label="ML Student")

plt.hist(py_students_age,bins,rwidth=0.8,histtype="bar",
        orientation="vertical",color="y",label="PY Student")

plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

```python
 bins=[15,20,25,30,35,40,45]
plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,histtype="bar",
        orientation="vertical",color=["m","y"],label=["ML Student","PY Student"])



plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

```python
# adding some size enlarge 

bins=[15,20,25,30,35,40,45]

plt.figure(figsize=(16,9))
plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,histtype="bar",
        orientation="vertical",color=["m","y"],label=["ML Student","PY Student"])



plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

```python
# to change the style of the graph using style package

from matplotlib import style
style.use("ggplot")

bins=[15,20,25,30,35,40,45]

plt.figure(figsize=(16,9))
plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,histtype="bar",
        orientation="vertical",color=["m","y"],label=["ML Student","PY Student"])



plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```
